```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_path = 'D:\coding\Pune_weather_forcasting\src\data\pune.csv'
```

```python
df = pd.read_csv(df_path,encoding='utf-8')

df.head()
```

```python
df.info()
```

```python
num_features = df.select_dtypes(include=['int64','float64'])
print(num_features.columns)
```

```python
cat_features = df.select_dtypes(include=object).drop(columns=['date_time'])
cat_features.head()
```

gonna have to convert these to date_time objects to get numerical data so that we can plot it


```python
ax = plt.axes()
sns.heatmap(df.isna(),cbar=False,ax=ax)
```

```python
df.dropna(axis=0)
```

```python
for col_name in cat_features.columns:
    no_name = "No " + df[col_name]
    df[col_name] = df[col_name].where(~df[col_name].str.startswith("No "), pd.NA)
    df[col_name] = pd.to_datetime(df[col_name],format = "%I:%M %p",errors="coerce")
    min_name = col_name+"_minutes"
    df[min_name] = df[col_name].dt.hour*60 + df[col_name].dt.minute
```

```python
# df['totalSnow_cm']
df=df.drop(['totalSnow_cm'],axis=1)
```

```python
df=df.drop(['moonrise', 'moonset', 'sunrise', 'sunset'],axis=1)
df.columns
```

```python
sample_rate = 50
sampled_df = df.iloc[::sample_rate]
fig,ax= plt.subplots(1,4,figsize=(20,7))
plt.subplot(151)
plt.plot(sampled_df['date_time'],sampled_df['moonrise_minutes'])
plt.subplot(152)
plt.plot(sampled_df['date_time'],sampled_df['moonset_minutes'])
plt.subplot(153)
plt.plot(sampled_df['date_time'],sampled_df['sunrise_minutes'])
plt.subplot(154)
plt.plot(sampled_df['date_time'],sampled_df['sunset_minutes'])
```
